```python
param_grid = {
    'penalty': ['l1', 'l2'],
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'solver': ['liblinear']
}


model = LogisticRegression(max_iter=1000)

cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

smote = SMOTE(random_state=42)

scaler = StandardScaler()

independent_variables = ['age', 'height_cm', 'weight_kg', 'weak_foot', 'skill_moves', 'pace', 'shooting',
       'passing', 'dribbling', 'defending', 'physic', 'attacking_crossing',
       'attacking_finishing', 'attacking_heading_accuracy',
       'attacking_short_passing', 'attacking_volleys', 'skill_dribbling',
       'skill_curve', 'skill_fk_accuracy', 'skill_long_passing',
       'skill_ball_control', 'movement_acceleration', 'movement_sprint_speed',
       'movement_agility', 'movement_reactions', 'movement_balance',
       'power_shot_power', 'power_jumping', 'power_stamina', 'power_strength',
       'power_long_shots', 'mentality_aggression', 'mentality_interceptions',
       'mentality_positioning', 'mentality_vision', 'mentality_penalties',
       'mentality_composure', 'defending_marking_awareness',
       'defending_standing_tackle', 'defending_sliding_tackle',
       'preferred_foot_Left', 'preferred_foot_Right', 'Attacker', 'Midfielder',
       'Defender']

dependent_variable = ['big_potential']

X = df[independent_variables]
y = df[dependent_variable].values.ravel()

num_features_to_select = 15
rfe = RFE(LogisticRegression(max_iter=1000), n_features_to_select=num_features_to_select)

grid_search = GridSearchCV(model, param_grid, cv=cv, scoring='f1_macro')

classification_reports = []
confusion_matrices = []

all_y_test = []
all_y_pred = []

for i, (train_idx, test_idx) in enumerate(cv.split(X, y)):
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]

    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    X_train_rfe = rfe.fit_transform(X_train_scaled, y_train)
    X_test_rfe = rfe.transform(X_test_scaled)

    X_train_smote, y_train_smote = smote.fit_resample(X_train_rfe, y_train)

    grid_search.fit(X_train_smote, y_train_smote)

    best_model = grid_search.best_estimator_
    best_model.fit(X_train_smote, y_train_smote)

    y_pred = best_model.predict(X_test_rfe)

    all_y_test.extend(y_test)
    all_y_pred.extend(y_pred)

    classification_report_str = classification_report(y_test, y_pred, output_dict=True, digits=3)
    classification_reports.append(classification_report_str)

    matrix_fold = confusion_matrix(y_test, y_pred)
    confusion_matrices.append(f"Confusion Matrix - Fold {i + 1}:\n{matrix_fold}\n")

    print(f'Fold {i + 1} - Classification Report:\n{classification_report(y_test, y_pred)}')
    print('-' * 50)

overall_classification_report = classification_report(all_y_test, all_y_pred, output_dict=True, digits=3)
overall_confusion_matrix = confusion_matrix(all_y_test, all_y_pred)

overall_cohen_kappa = cohen_kappa_score(all_y_test, all_y_pred)
overall_balanced_accuracy = balanced_accuracy_score(all_y_test, all_y_pred)

print("Overall Classification Report:")
print(overall_classification_report)
print('-' * 50)
print("Overall Confusion Matrix:")
print(overall_confusion_matrix)
print('-' * 50)
print(f"Overall Cohen's Kappa Score: {overall_cohen_kappa:.3f}")
print(f"Overall Balanced Accuracy: {overall_balanced_accuracy:.3f}")

print("Best Hyperparameters:")
print(grid_search.best_params_)
```
