```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_recall_fscore_support, cohen_kappa_score, balanced_accuracy_score
from sklearn.model_selection import StratifiedKFold
from imblearn.over_sampling import SMOTE
import pandas as pd
import numpy as np

['age', 'attacking_crossing', 'attacking_finishing', 'attacking_short_passing', 'skill_dribbling', 'skill_curve', 'skill_long_passing', 'skill_ball_control', 
 'movement_agility', 'movement_balance', 'power_shot_power',  'power_strength', 'power_long_shots', 'mentality_vision', 'mentality_penalties', 'mentality_composure', 
 'defending_standing_tackle', 'defending_sliding_tackle']

target_variable = ["big_potential"]

X = df[independent_variables]
y = df[target_variable]

y = np.ravel(y)

classification_reports = []
kappa_scores = []
conf_matrices = []
balanced_accuracies = []  

all_y_true = []
all_y_pred = []

fixed_hyperparameters = {
    'n_estimators': 50,
    'max_depth': 100,
    'criterion': 'gini',
    'bootstrap': False
}

skf = StratifiedKFold(n_splits=10, random_state=42, shuffle=True)

for fold_num, (train_index, test_index) in enumerate(skf.split(X, y), 1):
    X_train, X_val = X.iloc[train_index], X.iloc[test_index]
    y_train, y_val = y[train_index], y[test_index]

    smote = SMOTE(random_state=42)
    X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

    rf_model = RandomForestClassifier(random_state=42, **fixed_hyperparameters)

    rf_model.fit(X_train_resampled, y_train_resampled)

    y_val_pred = rf_model.predict(X_val)

    all_y_true.extend(y_val)
    all_y_pred.extend(y_val_pred)

    accuracy_fold = accuracy_score(y_val, y_val_pred)
    precision_fold, recall_fold, f1_fold, _ = precision_recall_fscore_support(y_val, y_val_pred, average='binary')
    kappa_fold = cohen_kappa_score(y_val, y_val_pred)
    balanced_accuracy_fold = balanced_accuracy_score(y_val, y_val_pred)  # Calculate balanced accuracy

    report_fold = classification_report(y_val, y_val_pred, digits=3)
    classification_reports.append(f"Fold {fold_num}:\n{report_fold}\n")
    kappa_scores.append(kappa_fold)
    balanced_accuracies.append(balanced_accuracy_fold)
    conf_matrices.append(confusion_matrix(y_val, y_val_pred))

print("Combined Classification Reports for All Folds:\n")
print("\n".join(classification_reports))

print("Cohen's Kappa Scores for Each Fold:\n", kappa_scores)

print("Balanced Accuracies for Each Fold:\n", balanced_accuracies)

final_conf_matrix = confusion_matrix(all_y_true, all_y_pred)
print("Confusion Matrix for Final Evaluation on Entire Dataset:\n", final_conf_matrix)

final_report = classification_report(all_y_true, all_y_pred, digits=3)
final_kappa = cohen_kappa_score(all_y_true, all_y_pred)
final_balanced_accuracy = balanced_accuracy_score(all_y_true, all_y_pred)
print("Final Classification Report on Entire Dataset:\n", final_report)
print("Cohen's Kappa for Final Evaluation on Entire Dataset:", final_kappa)
print("Balanced Accuracy for Final Evaluation on Entire Dataset:", final_balanced_accuracy)

print("Fixed Hyperparameters:", fixed_hyperparameters)
```
